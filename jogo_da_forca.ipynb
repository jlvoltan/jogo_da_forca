```
#importação dos módulos
import random  
import palavras  #arquivo py
```

```
#função para sortear uma palavra da lista
def sortear_palavra():
    palavra_escolhida = random.choice(palavras.girias_militares)
    return palavra_escolhida
```

```
#função para desenhar o boneco na forca
def desenha_forca(tentativa):
    progresso = [ 
               # Vazio
                """
                   ==========
                   ||      _|_
                   ||     
                   ||       
                   ||       
                   ||        
                   
                """,
               # cabeça, 
                """
                   ==========
                   ||      _|_
                   ||     (  )
                   ||       
                   ||       
                   ||        
                   
                """,
               # cabeça, olhos, ,
                """
                   ==========
                   ||      _|_
                   ||     (º º)
                   ||       
                   ||       
                   ||        
                   
                """,
                 
               # cabeça, olhos, boca,
                """
                   ==========
                   ||      _|_
                   ||     (º_º)
                   ||       
                   ||       
                   ||        
                   
                """,
               # cabeça, olhos, boca, pescoço e tronco
                """
                   ==========
                   ||      _|_
                   ||     (º_º)
                   ||       |
                   ||       |
                   ||        
                   
                """,
               # cabeça, olhos, boca, pescoço e tronco, 1 braço 
                """
                   ==========
                   ||      _|_
                   ||     (º_º)
                   ||     __|
                   ||       |
                   ||        
                   
                """,
               # cabeça, olhos, boca, pescoço e tronco, braços 
                """
                   ==========
                   ||      _|_
                   ||     (º_º)
                   ||     __|__
                   ||       |
                   ||        
                   
                """,
               # cabeça, olhos, boca, pescoço e tronco, braços e 1 perna
                """
                   ==========
                   ||      _|_
                   ||     (º_º)
                   ||     __|__
                   ||       |
                   ||      / 
                   
                """,
                # cabeça, olhos, boca, pescoço e tronco, braços e pernas
                """
                   ==========
                   ||      _|_
                   ||     (º_º)
                   ||     __|__
                   ||       |
                   ||      / / 
                   
                """
    ]
    return progresso[tentativa]
```

```
#função para iniciar o jogo
def jogar(palavra_escolhida):
    acertou = False
    erros = 0

    palavra_escolhida = palavra_escolhida.lower()
    print("="*30)
    print("BEM VINDO AO JOGO DA FORCA")
    print("="*30)

    letra = "*" * len(palavra_escolhida)
    
    lista_de_letras = []
    lista_de_palavras = []
    

    print("Let's Play!")
    print(desenha_forca(erros))
    print(letra)
    print("\n")

    while not acertou and erros < 8:
        tentativa = input("Por favor, informe uma letra ou tente uma palavra: ").lower()
        if len(tentativa) == 1 :                                                  #Caso em que escolheu uma letra
            if tentativa in lista_de_letras:                                      #Já tentou essa letra antes
                print("Você já tentou essa letra antes:", tentativa)
            elif tentativa not in palavra_escolhida:                              #Errou...
                print("A palavra não possui essa letra:", tentativa)
                erros += 1
                lista_de_letras.append(tentativa)
            else:
                print("Boa, a palavra tem a letra ", tentativa, " !")
                lista_de_letras.append(tentativa)
                posicao_letra = list(letra)  
                indices=[]
                posicao=0
                while posicao <len(palavra_escolhida): #encontrar as posições da letra certa
                  if(palavra_escolhida[posicao] == tentativa): indices.append(posicao)
                  posicao+=1
                for index in indices:
                    posicao_letra[index] = tentativa   #preenchendo as posições * com a letra
                letra = "".join(posicao_letra)
                if "*" not in letra:                      #Fim de jogo, temos um Campeão
                    acertou = True                                                                                 
        elif (len(tentativa) == len(palavra_escolhida)):
            if tentativa in lista_de_palavras:
                print("Você já tentou essa palavra: ", tentativa)
            elif tentativa == palavra_escolhida:
                acertou = True
                letra = palavra_escolhida
            else:
                print("Será?? ", tentativa, " - ERROU!!! ")
                erros += 1
                lista_de_palavras.append(tentativa)
               
        else:
            print("Passou longe!!! ")
        print(desenha_forca(erros))
        print(letra)
        print("\n")
    if acertou:
        print("Parabéns!! Palavra correta !! ",  tentativa)
    else:
        print(" Forca! Forca! Forca! Você errou, a palavra era " + palavra_escolhida + ". Tente outra vez!")
        print("\n")
```
